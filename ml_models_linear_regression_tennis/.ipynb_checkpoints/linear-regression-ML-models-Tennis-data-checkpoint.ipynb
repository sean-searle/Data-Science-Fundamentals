# Building Linear Regression ML Models

In this project, I build single and multiple feature linear regression models, check their performance, and compare the model outputs to actual outcome data to evaluate accuracy. The purpose of this project is to pratice creating and using simple and multiple variable linear regression ML models with Scikit-Learn.

The data is taken from the [Association of Tennis Professionals (ATP)](https://datahub.io/sports-data/atp-world-tour-tennis-data). My aim is to use this data to train linear regression models that predict the tournament outcomes for tennis players based on their playing habits and statistics.

### Setup

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### Loading and investigating the data

```python
df = pd.read_csv("tennis_stats.csv")
# print(df.head())
print(df.info())
```

### Exploratory data analysis (EDA)

```python
plt.scatter(df.Aces, df.Wins, alpha=.4)
plt.xlabel("Aces")
plt.ylabel("Wins")
plt.title("Wins vs Aces")
```

```python
plt.scatter(df.BreakPointsFaced, df.Losses, alpha=.4)
plt.xlabel("Break Points Faced")
plt.ylabel("Losses")
plt.title("Losses vs Break Points Faced")
```

```python
plt.scatter(df.BreakPointsOpportunities, df.Wins, alpha=.4)
plt.xlabel("Break Points Opportunities")
plt.ylabel("Wins")
plt.title("Wins vs Break Points Opportunities")
```

### EDA conclusions

There are strong positive correlations between these pairs of features, especially Losses vs Break Points Faced and Wins vs Break Point Opportunities. These seem to predict Losses and Wins, respectively.

### Simple linear regressions

To start, I selected the "break points faced" feature and the "losses" outcome to build a single feature linear regression model. The code below isolates those variables.

```python
break_pts_faced_x = df.BreakPointsFaced
losses_y = df.Losses
print(losses_y.head())
```

Creating an 80/20 train/test split from the data and turning the resulting series into reshaped arrays.

```python
x_train, x_test, y_train, y_test = train_test_split(break_pts_faced_x, losses_y, train_size=0.8, \
                                                    test_size=0.2, random_state=6)
x_train = np.array(x_train).reshape(-1, 1)
y_train = np.array(y_train).reshape(-1, 1)
x_test = np.array(x_test).reshape(-1, 1)
y_test = np.array(y_test).reshape(-1, 1)
```

Creating the model and fitting the data.

```python
lrm = LinearRegression()
lrm.fit(x_train, y_train)
```

Checking models accuracy with the .score() method. Ideally, the model scores over 0.70. Finding the predicted outcomes based on my model and comparing them to the actual outcome data with plots.

```python
print(lrm.score(x_train, y_train))
print(lrm.score(x_test, y_test))

prediction = lrm.predict(x_test)
plt.scatter(y_test,prediction, alpha=0.4)
plt.title("Predicted vs Actual Losses")
plt.xlabel("Actual Losses")
plt.ylabel("Predicted Losses")
```

```python
plt.scatter(df.BreakPointsFaced, df.Losses, alpha=.4)
plt.plot(x_test, prediction)
plt.title("Actual Data w/ Regression Line")
plt.xlabel("Break Points Faced")
plt.ylabel("Losses")
plt.show()
```

Repeating this process for the "Aces" feature and "Wins" Outcome.

```python
aces_x = df.Aces
wins_y = df.Wins

x_train2, x_test2, y_train2, y_test2 = train_test_split(aces_x, wins_y, train_size=0.8, \
                                                    test_size=0.2, random_state=6)
x_train2 = np.array(x_train2).reshape(-1, 1)
y_train2 = np.array(y_train2).reshape(-1, 1)
x_test2 = np.array(x_test2).reshape(-1, 1)
y_test2 = np.array(y_test2).reshape(-1, 1)

lrm2 = LinearRegression()
lrm2.fit(x_train2, y_train2)
```

```python
print(lrm2.score(x_test2, y_test2))

prediction2 = lrm2.predict(x_test2)
plt.scatter(y_test2,prediction2, alpha=0.4)
plt.title("Predicted vs Actual Wins")
plt.xlabel("Actual Wins")
plt.ylabel("Predicted Wins")
```

### Single feature linear regressions conclusions

Through Residual Analysis, I saw that the linear regressions model predicting the Losses outcome based on the Break Points Faced feature was fairly accurate. The model predicting Wins based on Aces was far less accurate and produced a coefficient of determination (R^2) of less that .70. Initially, the Wins vs Aces pairing did look less correlated than the other features I explored in the EDA section.

### Multiple feature linear regressions

I'm interested in whether the "Return Games Won" and "Break Points Opportunities" features can be used to create an accruate linear regression model to predict "Winnings".

The code below selects features and outcomes to predict and train/test splits the data.

```python
features_x = df[['ReturnGamesWon', 'BreakPointsOpportunities']]
winngings_y = df.Winnings

x_train3, x_test3, y_train3, y_test3 = train_test_split(features_x, winngings_y, train_size=0.8, \
                                                    test_size=0.2, random_state=6)
```

Create and train model on training data.

```python
model = LinearRegression()
model.fit(x_train3, y_train3)
```

Score model on test data.

```python
print(model.score(x_test3, y_test3))
```

Make predictions with model and plot predictions against actual winnings.

```python
winnings_prediction = model.predict(x_test3)

plt.scatter(y_test3, winnings_prediction, alpha=.4)
plt.title("Predicted vs Actual Winnings - 2 Features")
plt.xlabel("Actual Winnings")
plt.ylabel("Predicted Winnings")
```

Now I will create a multiple feature model with many more features.

```python
features = df[['FirstServe','FirstServePointsWon','FirstServeReturnPointsWon',
'SecondServePointsWon','SecondServeReturnPointsWon','Aces',
'BreakPointsConverted','BreakPointsFaced','BreakPointsOpportunities',
'BreakPointsSaved','DoubleFaults','ReturnGamesPlayed','ReturnGamesWon',
'ReturnPointsWon','ServiceGamesPlayed','ServiceGamesWon','TotalPointsWon',
'TotalServicePointsWon']]
outcome = df[['Winnings']]

x_train4, x_test4, y_train4, y_test4 = train_test_split(features, outcome, train_size=0.8, \
                                                    test_size=0.2, random_state=6)
```

```python
model = LinearRegression()
model.fit(x_train4, y_train4)
```

```python
print(model.score(x_test4, y_test4))
```

```python
winnings_prediction2 = model.predict(x_test4)

plt.scatter(y_test4, winnings_prediction2, alpha=.4)
plt.title("Predicted vs Actual Winnings - Many Features")
plt.xlabel("Actual Winnings")
plt.ylabel("Predicted Winnings")
```

### Multiple feature linear regressions conclusions

The multiple feature model using many more features (18 instead of 2) was more accurate. It produced a higher coefficient of determination (accuracy score) and the scatter plot of the model prediction data vs actual outcome data had a tighter spread.
